## Task 2

### Build a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
pred=model.predict(X_val)
pred1=model.predict(X_train)
```

### Evaluate your model with Mean Squared Error (MSE)

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y_val,pred) #error on val data
```

```python
mean_squared_error(Y_train,pred1) # error on train data
```

### Build Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
model2=RandomForestRegressor()
model2.fit(X_train,Y_train)
pred2=model2.predict(X_val)
pred12=model2.predict(X_train)
```

### Evaluate your model with Mean Squared Error (MSE)

```python
mean_squared_error(Y_val,pred2) #error on val data
```

```python
mean_squared_error(Y_train,pred12) #error on train data
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
model2 = RandomForestRegressor()
parameters = {
    'n_estimators': [100,1000,5000,10000],
    'random_state': [1,None],
    'max_depth': [5,10]

}
rs = RandomizedSearchCV(model2, parameters,n_iter=10)
rs.fit(X_important_train, Y_train)
print_results(rs)
```
