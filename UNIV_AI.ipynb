```python
from keras.models import Sequential
from keras.layers import InputLayer,Dense
```

```python
from keras.layers import Dropout
```

```python
X_train.shape
```

```python
input_neurons=X_train.shape[1]
output_neurons=1
NHL_1=400
NHL_2=250
NHL_3=150
NHL_4=75
NHL_5=10
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(input_neurons,)))
model.add(Dense(units=NHL_1,activation='relu'))
model.add(Dense(units=NHL_2,activation='relu'))
model.add(Dense(units=NHL_3,activation='relu'))
model.add(Dense(units=NHL_4,activation='relu'))
model.add(Dense(units=NHL_5,activation='relu'))
model.add(Dense(units=output_neurons,activation='sigmoid'))
```

```python
model.summary()
```

```python
from keras.optimizers import Adam
optimizers=Adam()
```

```python
model.compile(loss='binary_crossentropy',optimizer=optimizers,metrics=['accuracy'])
```

```python
model_history=model.fit(X_train,Y_train,batch_size=128,epochs=10)
```
