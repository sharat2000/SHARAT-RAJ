```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train_data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_note_data/training_set_label.csv" )
```

```
iqr1=3.574450-(-1.574975)
q_up1=3.574450+1.5*iqr1
q_up1
```

```
iqr2=0.399980-(-2.246975)
q_low2=-2.246975-1.5*iqr2
q_low2
```

```
x=train_data[train_data['CWTI']<q_up1][train_data['EI']>q_low2]
x.count()
```

```
Xtrain=x.iloc[:,0:4]
Ytrain=x.iloc[:,4]
Xtrain.head()
```

```
Ytrain.head()
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn import svm
```

```
model = LogisticRegression()
model.fit(Xtrain, Ytrain)
model.score(Xtrain,Ytrain)
```

```

Y_test1 = model.predict(test_data)
Y_test1
```

```
model=svm.SVC(kernel='linear')
model.fit(Xtrain, Ytrain)
model.score(Xtrain,Ytrain)
```

```
Y_test2 = model.predict(test_data)
Y_test2
```

```
model=RandomForestClassifier()
model.fit(Xtrain, Ytrain)
model.score(Xtrain,Ytrain)
```

```
Y_test3 = model.predict(test_data)
Y_test3
```
